```python
# Import necessary libraries
from sklearn.cluster import DBSCAN
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd

# Load Iris dataset
iris = load_iris()
X = iris.data  # Use all features
y = iris.target

# Fit DBSCAN model
dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters = dbscan.fit_predict(X)

# Create a DataFrame for plotting
df = pd.DataFrame(iris.data[:, :2], columns=['Sepal Length', 'Sepal Width'])  # Use only first two features for visualization
df['Cluster'] = clusters

# Plot the DBSCAN clusters
plt.figure(figsize=(8,6))
plt.scatter(df['Sepal Length'], df['Sepal Width'], c=df['Cluster'], cmap='plasma', s=50)
plt.title('DBSCAN Clustering on Iris Dataset (Visualized with Sepal Length vs Sepal Width)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.colorbar(label='Cluster')
plt.show()
```
