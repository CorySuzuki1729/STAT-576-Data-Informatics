```python
# Import all necessary libraries.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.manifold import MDS, Isomap, LocallyLinearEmbedding, TSNE
from adjustText import adjust_text
```

```python
# Load the data and make the target variable. I used the head() functions to get glimpses of what each of my
# variables and data looked like prior to algorithm fitting/transformations. We use sample size n = 200. 

mnist_data = pd.read_csv("C://Users/coryg/Onedrive/Desktop/STAT_576_F24/mnist_hw.csv")
mnist_data.head()
mnist_data = mnist_data
mnist_data.head()

class_target = mnist_data.iloc[:, -1]
mnist_mat = mnist_data.iloc[:, 0:]
mnist_mat.head()
class_target.head()
```

```python
# Problem 1: Feature Extraction by MDS

mds_mnist_res = MDS(n_components=2, random_state=42).fit_transform(mnist_mat)
X_sample1 = np.random.choice(len(mds_mnist_res), size = 200, replace = False)
mds_sample = mds_mnist_res[X_sample1]
y_sample1 = class_target.iloc[X_sample1]
```

```python
# Problem 2: Feature Extraction by IsoMAP

isomap_mnist_res = Isomap(n_neighbors=10, n_components=2).fit_transform(mnist_mat)
X_sample2 = np.random.choice(len(isomap_mnist_res), size = 200, replace = False)
isomap_sample = isomap_mnist_res[X_sample2]
y_sample2 = class_target.iloc[X_sample2]
```

```python
# Problem 3: Feature Extraction by LLE

lle_mnist_res = LocallyLinearEmbedding(n_neighbors=10, n_components=2).fit_transform(mnist_mat)
X_sample3 = np.random.choice(len(lle_mnist_res), size = 200, replace = False)
lle_sample = lle_mnist_res[X_sample3]
y_sample3 = class_target.iloc[X_sample3]
```

```python
# Problem 4: Feature Extraction bt t-SNE

tsne_mnist_res = TSNE(n_components=2, perplexity=5, random_state=0).fit_transform(mnist_mat)
X_sample4 = np.random.choice(len(tsne_mnist_res), size = 200, replace = False)
tsne_sample = tsne_mnist_res[X_sample4]
y_sample4 = class_target.iloc[X_sample4]
```

```python
def plot_scatter(data, sample_target, method_name):
    label_encoder = LabelEncoder()
    target_encoded = label_encoder.fit_transform(sample_target)
    plt.figure(figsize=(10, 7))
    scatter = plt.scatter(data[:, 0], data[:, 1], c=target_encoded, cmap='viridis')
    plt.title(f'{method_name} Scatter Plot with First Two Components')
    plt.xlabel(f'First {method_name} Dimension')
    plt.ylabel(f'Second {method_name} Dimension')

    text_list = [plt.text(data[i, 0], data[i, 1], str(sample_target.iloc[i]), fontsize=9) for i in range(len(sample_target))]
    adjust_text(text_list, arrowprops=dict(arrowstyle='-', color='gray', lw=0.5))
    plt.colorbar(scatter)
    plt.show()
```

```python
plot_scatter(mds_sample, y_sample1, 'MDS')
plot_scatter(isomap_sample, y_sample2, 'Isomap')
plot_scatter(lle_sample, y_sample3, 'LLE')
plot_scatter(tsne_sample, y_sample4, 't-SNE')
```
